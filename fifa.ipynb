```python
import pandas as pd
fifaDT = pd.read_csv('./fifa21_raw_data.csv')
pd.set_option('display.max_columns',85)
pd.set_option('display.max_rows', 85)
```

```python
fifaDT.dtypes
```

```python
fifaDT.rename(columns={'Weight':'Weight (lbs)','Height':'Height (ft)'},inplace=True)
fifaDT['Weight (lbs)'] = fifaDT['Weight (lbs)'].apply(lambda x: x.replace('lbs',''))
fifaDT['Weight (lbs)'] = fifaDT['Weight (lbs)'].astype(int)
fifaDT['Weight (lbs)']
# Renomeando a coluna Weight para algo mais específico, retirando strings desnecessários desta coluna  (lbs)
# e mudando seu tipo para inteiro
# Transformação de dados
```

```python
def removerAspas(dado):
    NovoDado = dado.replace('"','')
    NovoDado = NovoDado.replace("'",'.')
    return NovoDado
fifaDT['Height (ft)'] = fifaDT['Height (ft)'].apply(removerAspas)
fifaDT['Height (ft)'] = fifaDT['Height (ft)'].astype(float)
fifaDT['Height (ft)']

# Mais Transformação de dados, retirando caracteres desnecessários (" e ') e mudando o tipo para FLOAT
```

```python
def formatarHits(numero):
    NovoDado = numero.replace('\n','')
    NovoDado = NovoDado.replace('.','')
    NovoDado = NovoDado.replace('K','00')
    return NovoDado
fifaDT['Hits'] = fifaDT['Hits'].astype(str)
fifaDT['Hits'] = fifaDT['Hits'].apply(formatarHits)
fifaDT['Hits'] = fifaDT['Hits'].astype(int)
fifaDT['Hits']
```

```python
import calendar
fifaDT['Joined'] = pd.to_datetime(fifaDT['Joined'])
fifaDT['Month'] = fifaDT['Joined'].dt.month.apply(lambda x: calendar.month_name[x])
fifaDT['Date'] = fifaDT['Joined'].dt.day
fifaDT['Year'] = fifaDT['Joined'].dt.year
fifaDT[['Month','Date','Year']]
```

```python
def formatarNum(numero):
    NovoDado = numero.replace('€','')
    NovoDado = NovoDado.replace('.','')
    NovoDado = NovoDado.replace('K','00')
    NovoDado = NovoDado.replace('M','00000')
    return NovoDado
fifaDT.rename(columns={'Value':'Value (€)','Wage':'Wage (€)','Release Clause':'Release Clause (€)'},inplace=True)
fifaDT['Value (€)'] = fifaDT['Value (€)'].apply(formatarNum)
fifaDT['Wage (€)'] = fifaDT['Wage (€)'].apply(formatarNum)
fifaDT['Release Clause (€)'] = fifaDT['Release Clause (€)'].apply(formatarNum)
fifaDT[['Value (€)','Wage (€)','Release Clause (€)']]
```

```python
fifaDT
```

```python
def formatarTeamContract(texto):
    NovoTexto = texto.replace('\n',' ')
    NovoTexto = NovoTexto.replace('~',' ')
    NovoTexto = NovoTexto.strip()
    return NovoTexto
fifaDT['Team & Contract'] = fifaDT['Team & Contract'].apply(formatarTeamContract)
fifaDT['Team & Contract'] = fifaDT['Team & Contract'].str.replace(r'\s+', ' ',regex=True) # Remove longos espaços
fifaDT[['Team','ContractStart','ContractEnd']] = fifaDT['Team & Contract'].str.rsplit(' ',n=2,expand=True) # Separa a series Team & Contract em 3 colunas, baseando essa divisão no 3° espaço da direita pra esquerda
fifaDT['Contract'] = fifaDT['ContractStart'] + ' ' + fifaDT['ContractEnd'] # Cria a coluna Contract
fifaDT['Contract'] = fifaDT['Contract'].str.replace(' ','~') # Recoloca o símbolo de ~
fifaDT = fifaDT.drop('ContractStart', axis=1) # Remove as colunas criadas temporariamente
fifaDT = fifaDT.drop('ContractEnd', axis=1)
fifaDT['Contract']
```
